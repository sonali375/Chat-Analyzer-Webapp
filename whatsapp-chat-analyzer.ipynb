```python
import re
import pandas as pd
import datetime
```

```python
f=open('grp_chat.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern="\d{2}\/\d{2}\/\d{4},\s\d{1,2}:\d{1,2}:\d{1,2}\s[A-Z][A-Z]\s-\s"
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
#datetime_formats = {
 #       '12hr' : '%d/%m/%Y, %H:%M:%S %p - ',
  #      '24hr' : '%d/%m/%Y, %H:%M:%S - ',
   #     'custom': ''
#}
```

```python
df = pd.DataFrame({'date_time': dates, 'user_msg': messages})
df['date_time'] = pd.to_datetime(df['date_time'], format='%d/%m/%Y, %H:%M:%S %p - ')
```

```python
df.shape
```

```python
df
```

```python
#separate users and their messages
usernames = []
msgs = []
for i in df['user_msg']:
    a = re.split('([\w\W]+?):\s', i) 
    if(a[1:]): 
        usernames.append(a[1])
        msgs.append(a[2])
    else: 
        usernames.append("group_notification")
        msgs.append(a[0])

             
df['user'] = usernames
df['message'] = msgs

    
df.drop('user_msg', axis=1, inplace=True)
```

```python
f = open('Hinglish stopwords.txt','r')
stop_words = f.read()
print(stop_words)
```

```python
words=[]
for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
period=[]
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str("00"))
    elif hour == 0:
        period.append(str("00") + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
period
```

```python
df['period'] = period
```
